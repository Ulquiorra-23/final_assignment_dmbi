```python
#SALES DATAFRAME STATUS OVERVIEW

print(sales_phases_funnel_df.isnull().sum())  # Count missing values per column

print('-------------------------')
print(sales_phases_funnel_df.info()) #info
```

```python
# CLEANING SALES FUNNEL DATAFRAME - 1

#DROPPING DUPLICATES

#count duplicates

print('There are ', sales_phases_funnel_df.duplicated().sum(), ' duplicate rows before duplicate cleaning') 

sales_phases_funnel_df.drop_duplicates(inplace=True)

print('There are ', sales_phases_funnel_df.duplicated().sum(), ' duplicate rows after duplicate cleaning') 
```

```python
# CLEANING SALES FUNNEL DATAFRAME -2

# DELETE UNUSABLE LEADS

# Drop rows where KO_REASON is "Unreachable"
sales_phases_funnel_df = sales_phases_funnel_df[~((sales_phases_funnel_df['CURRENT_PHASE'] == 'KO') & (sales_phases_funnel_df['KO_REASON'] == 'Unreachable'))]

# Reset the index of the updated DataFrame
sales_phases_funnel_df.reset_index(drop=True, inplace=True)

# Verify the changes
#print(sales_phases_funnel_df.isnull().sum())
#print(sales_phases_funnel_df.head())

sales_phases_funnel_df
```

```python
# CLEANING SALES FUNNEL DATAFRAME - 3

# REMOVE OUTLIERS

# Calculate Q1 (25th percentile) and Q3 (75th percentile)
Q1 = sales_phases_funnel_df['INSTALLATION_PRICE'].quantile(0.25)
Q3 = sales_phases_funnel_df['INSTALLATION_PRICE'].quantile(0.75)

# Calculate the Interquartile Range (IQR)
IQR = Q3 - Q1

# Identify outliers inside a new Data Frame
outliers_df = sales_phases_funnel_df[(sales_phases_funnel_df['INSTALLATION_PRICE'] < (Q1 - 1.5 * IQR)) | 
                                     (sales_phases_funnel_df['INSTALLATION_PRICE'] > (Q3 + 1.5 * IQR))]

# Print the number of outliers
print(f"Number of outliers: {len(outliers_df)}")

# Update sales_phases_funnel_df to exclude the outliers
sales_phases_funnel_df = sales_phases_funnel_df[~((sales_phases_funnel_df['INSTALLATION_PRICE'] < (Q1 - 1.5 * IQR)) | 
                                                  (sales_phases_funnel_df['INSTALLATION_PRICE'] > (Q3 + 1.5 * IQR)))]

# Print the DataFrame without outliers
print("DataFrame without outliers:")
print(sales_phases_funnel_df)

# Print the DataFrame with only outliers
print("DataFrame with only outliers:")
print(outliers_df)


sales_phases_funnel_df
```

```python
# CLEANING SALES FUNNEL DATAFRAME - 4

# FORCING DATA TYPES

###### this should be a string
####instead of doing this
####declare data types at the beginning and read them with read_csv

# Changing ZIPCODE column data type to interger as zipcode don't have decimals
sales_phases_funnel_df['ZIPCODE'] = sales_phases_funnel_df['ZIPCODE'].astype('Int64')

print(sales_phases_funnel_df.info())


nan_zipcode_rows = sales_phases_funnel_df[sales_phases_funnel_df['ZIPCODE'].isnull()]

sales_phases_funnel_df
```
